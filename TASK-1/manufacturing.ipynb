# LR

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

# Predictions
y_train_pred = lr_model.predict(X_train)
y_test_pred = lr_model.predict(X_test)

# Evaluate model
train_r2 = r2_score(y_train, y_train_pred)
test_r2 = r2_score(y_test, y_test_pred)
print("Linear Regression Results:")
print("--------------------------")
print(f" R² Score: {train_r2:.4f}")
print(f"Test R² Score: {test_r2:.4f}")
print(f"Training RMSE: {train_rmse:.4f}")
print(f"Test RMSE: {test_rmse:.4f}")
# Feature importance (absolute coefficient values)
feature_importance = pd.Series(np.abs(lr_model.coef_), index=X.columns).sort_values(ascending=False)

print("\nTop 10 Important Features:")
print(feature_importance.head(10))

# Visualization of feature importance
plt.figure(figsize=(10, 6))
sns.barplot(x=feature_importance.head(10), y=feature_importance.head(10).index, palette='viridis')
plt.title("Top 10 Feature Importances (Linear Regression Coefficients)")
plt.xlabel("Coefficient Magnitude (Importance)")
plt.ylabel("Features")
plt.show()

```


-> The reasons why it is not predicting the feature improtance properly means:

* ("The lr assumes every fetaure is affects the target linealry")

1. some columns contain larger numeric values → so its coefficient becomes small, while smaller-ranged features may show bigger coefficients —
even if they are not actually more important because atlast model need to predict close to the actual not the cofficeints small or big.

2. If your data has non-linear patterns or Linear Regression cant capture them properly — so even important features may show small coefficients. where as tree can handle the non linearity and give feature importance.

3. When independent fetaures are corelated the model may split their coefficients — giving smaller weights then features look less important to the model. even if both are important. so solution is by combining the fetaures in to new feature or drop one feature.

# DT

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Train Decision Tree with Regularization
dt_model = DecisionTreeRegressor(
    random_state=42,
    max_depth=8,              # limits the depth of the tree
    min_samples_split=10,     # requires at least 10 samples to split a node
    min_samples_leaf=5,       # requires at least 5 samples per leaf
    max_features='sqrt'       # considers only a subset of features at each split
)
dt_model.fit(X_train, y_train)

# Step 2: Predictions
y_train_pred = dt_model.predict(X_train)
y_test_pred = dt_model.predict(X_test)

# Step 3: Evaluate Model
train_r2 = r2_score(y_train, y_train_pred)
test_r2 = r2_score(y_test, y_test_pred)
train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))

print("===== Decision Tree Regression (Regularized) =====")
print(f"Training R² Score: {train_r2:.4f}")
print(f"Test R² Score: {test_r2:.4f}")
print(f"Training RMSE: {train_rmse:.4f}")
print(f"Test RMSE: {test_rmse:.4f}")

# Step 4: Feature Importance
feature_importance = pd.Series(dt_model.feature_importances_, index=X.columns).sort_values(ascending=False)

print("\nTop 10 Important Features:")
print(feature_importance.head(10))

# Step 5: Visualization
plt.figure(figsize=(10, 6))
sns.barplot(x=feature_importance.head(10), y=feature_importance.head(10).index, palette='viridis')
plt.title("Top 10 Feature Importances (Decision Tree - Regularized)")
plt.xlabel("Importance Score")
plt.ylabel("Features")
plt.show()
```

# RF

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Step 1: Regularized Random Forest Model
rf_model = RandomForestRegressor(
    n_estimators=200,        # more trees for stability
    max_depth=10,            # limit tree depth
    min_samples_split=10,    # minimum samples to split an internal node
    min_samples_leaf=5,      # minimum samples at a leaf node
    max_features='sqrt',     # use subset of features for diversity
    bootstrap=True,          # sampling with replacement
    random_state=42,
    n_jobs=-1
)

# Step 2: Train Model
rf_model.fit(X_train, y_train)

# Step 3: Predictions
y_train_pred = rf_model.predict(X_train)
y_test_pred = rf_model.predict(X_test)

# Step 4: Evaluation
train_r2 = r2_score(y_train, y_train_pred)
test_r2 = r2_score(y_test, y_test_pred)
train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))

print("===== Regularized Random Forest Regression =====")
print(f"Training R² Score: {train_r2:.4f}")
print(f"Test R² Score: {test_r2:.4f}")
print(f"Training RMSE: {train_rmse:.4f}")
print(f"Test RMSE: {test_rmse:.4f}")

# Step 5: Feature Importance
feature_importance = pd.Series(rf_model.feature_importances_, index=X.columns).sort_values(ascending=False)

print("\nTop 10 Important Features:")
print(feature_importance.head(10))

# Step 6: Visualization
plt.figure(figsize=(10, 6))
sns.barplot(x=feature_importance.head(10), y=feature_importance.head(10).index, palette='viridis')
plt.title("Top 10 Feature Importances (Regularized Random Forest)")
plt.xlabel("Importance Score")
plt.ylabel("Features")
plt.show()
```

# OUTLERS

```python
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Loop through each column and clip outliers
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[col] = df[col].clip(lower=lower, upper=upper)
    print(f"Outliers in '{col}' capped between [{lower:.2f}, {upper:.2f}]")

# Step 3: Plot all numeric columns after clipping (in one grid)
plt.figure(figsize=(15, 8))
df[numeric_cols].boxplot()
plt.title("Boxplots After Clipping Outliers for All Numeric Columns")
plt.xticks(rotation=45)
plt.show()
```

# DT

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Step 1: Train Decision Tree with Regularization
dt_model = DecisionTreeRegressor(
    random_state=42,
    max_depth=8,              # limits the depth of the tree
    min_samples_split=10,     # requires at least 10 samples to split a node
    min_samples_leaf=5,       # requires at least 5 samples per leaf
    max_features='sqrt'       # considers only a subset of features at each split
)
dt_model.fit(X_train, y_train)

# Step 2: Predictions
y_train_pred = dt_model.predict(X_train)
y_test_pred = dt_model.predict(X_test)

# Step 3: Evaluate Model
train_r2 = r2_score(y_train, y_train_pred)
test_r2 = r2_score(y_test, y_test_pred)
train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))

print("===== Decision Tree Regression (Regularized) =====")
print(f"R² Score: {train_r2:.4f}")
print(f"Test R² Score: {test_r2:.4f}")
print(f"Training RMSE: {train_rmse:.4f}")
print(f"Test RMSE: {test_rmse:.4f}")
# Step 4: Feature Importance
feature_importance = pd.Series(dt_model.feature_importances_, index=X.columns).sort_values(ascending=False)

print("\nTop 10 Important Features:")
print(feature_importance.head(10))

# Step 5: Visualization
plt.figure(figsize=(10, 6))
sns.barplot(x=feature_importance.head(10), y=feature_importance.head(10).index, palette='viridis')
plt.title("Top 10 Feature Importances (Decision Tree - Regularized)")
plt.xlabel("Importance Score")
plt.ylabel("Features")
plt.show()
```

# RF

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Step 1: Regularized Random Forest Model
rf_model = RandomForestRegressor(
    n_estimators=200,        # more trees for stability
    max_depth=10,            # limit tree depth
    min_samples_split=10,    # minimum samples to split an internal node
    min_samples_leaf=5,      # minimum samples at a leaf node
    max_features='sqrt',     # use subset of features for diversity
    bootstrap=True,          # sampling with replacement
    random_state=42,
    n_jobs=-1
)

# Step 2: Train Model
rf_model.fit(X_train, y_train)

# Step 3: Predictions
y_train_pred = rf_model.predict(X_train)
y_test_pred = rf_model.predict(X_test)

# Step 4: Evaluation
train_r2 = r2_score(y_train, y_train_pred)
test_r2 = r2_score(y_test, y_test_pred)
train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))

print("===== Regularized Random Forest Regression =====")
print(f"Training R² Score: {train_r2:.4f}")
print(f"Test R² Score: {test_r2:.4f}")
print(f"Training RMSE: {train_rmse:.4f}")
print(f"Test RMSE: {test_rmse:.4f}")

# Step 5: Feature Importance
feature_importance = pd.Series(rf_model.feature_importances_, index=X.columns).sort_values(ascending=False)

print("\nTop 10 Important Features:")
print(feature_importance.head(10))

# Step 6: Visualization
plt.figure(figsize=(10, 6))
sns.barplot(x=feature_importance.head(10), y=feature_importance.head(10).index, palette='viridis')
plt.title("Top 10 Feature Importances (Regularized Random Forest)")
plt.xlabel("Importance Score")
plt.ylabel("Features")
plt.show()
```
